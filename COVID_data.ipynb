```r
bmi_df = read.csv("NCD_RisC_Lancet_2017_BMI_age_standardised_country.csv")
maxYear =  toString((max(bmi_df$Year)))
bmi_df <- bmi_df %>% filter(Year == maxYear)  %>% group_by(Entity, Code) %>% summarize(Mean_bmi = mean(Mean_BMI))
```

```r
bloodp_df <- read.csv("NCD_RisC_Lancet_2016_BP_age_standardised_countries.csv")
bloodp_df <- na.omit(bloodp_df) %>% mutate( mean_bloodPressure = as.numeric(Mean_systolic_blood_pressure_.mmHg.))
```

```r
maxYear_bp =  toString((max(bloodp_df$Year)))

bloodp_df <- na.omit(bloodp_df) %>% filter(Year == maxYear_bp) %>% group_by(Entity, Code) %>% summarize(Mean_bp = mean(mean_bloodPressure))
```

```r
df2 <- inner_join( df1, bloodp_df)
```
